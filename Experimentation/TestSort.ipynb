```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Load CSV files containing sorting algorithm performance data
resultats_merge_sort = pd.read_csv('merge_sort_results.csv')
resultats_onlogn = pd.read_csv('resultats_Tri_ONLOGN.csv')
resultats_gnome_sort = pd.read_csv('gnome_sort_results.csv')
resultats_quick_sort = pd.read_csv('quick_sort_results.csv')
```

```python
# Data Processing and Organization

# Extract execution times and input sizes from each CSV file
merge_sort_times = resultats_merge_sort['Temps de Tri (s)']
merge_sort_sizes = resultats_merge_sort['Taille']

onlogn_times = resultats_onlogn['Temps de Tri (s)']
onlogn_sizes = resultats_onlogn['Taille']

gnome_sort_times = resultats_gnome_sort['Temps de Tri (s)']
gnome_sort_sizes = resultats_gnome_sort['Taille']

quick_sort_times = resultats_quick_sort['Temps de Tri (s)']
quick_sort_sizes = resultats_quick_sort['Taille']

# Organize data into a dictionary for easier plotting
data = {
    'Merge Sort': (merge_sort_sizes, merge_sort_times),
    'O(N log N) Heap Sort': (onlogn_sizes, onlogn_times),
    'Gnome Sort': (gnome_sort_sizes, gnome_sort_times),
    'Quick Sort': (quick_sort_sizes, quick_sort_times)
}
```

```python
# Create Main Figure with Subplots

# Set up a figure with two subplots: one for main algorithms and one for Gnome Sort
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Plot main algorithms in the first subplot
for algorithm, (sizes, times) in data.items():
    if algorithm != 'Gnome Sort':
        ax1.plot(sizes, times, label=algorithm)

ax1.set_title('Sorting Algorithms (excluding Gnome Sort)')
ax1.set_xlabel('Input Size')
ax1.set_ylabel('Execution Time (s)')
ax1.legend()
ax1.grid(True)

# Plot Gnome Sort in the second subplot
gnome_sizes, gnome_times = data['Gnome Sort']
ax2.plot(gnome_sizes, gnome_times, label='Gnome Sort', color='orange')

ax2.set_title('Gnome Sort')
ax2.set_xlabel('Input Size')
ax2.set_ylabel('Execution Time (s)')
ax2.legend()
ax2.grid(True)

# Display the plots
plt.tight_layout()
plt.show()
```
