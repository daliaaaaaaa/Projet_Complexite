```python
import pandas  as pd
import matplotlib.pyplot  as plt 
```

```python
# Load the CSV files into dataframes
data_max_heap = pd.read_csv('complexite_temps_max_heap.csv')
data_min_heap = pd.read_csv('complexite_temps_min_heap.csv')
data_btree = pd.read_csv('complexite_temps_btree.csv')
data_list_bd = pd.read_csv('complexite_temps_bd_exp.csv')
data_abr = pd.read_csv('complexite_temps_abr.csv')
plt.figure(figsize=(18, 6))
```

```python
# Assuming the necessary calculations for temps_recherche_* and temps_suppression_* are done here
temps_recherche_abr = data_abr["Temps recherche"]
temps_recherche_list_bd = data_list_bd["Temps recherche (ms)"]
temps_recherche_btree = data_btree["Temps Recherche"]
temps_recherche_max_heap = data_max_heap["Temps Recherche"]

temps_suppression_abr = data_abr["Temps suppression"]
temps_suppression_list_bd = data_list_bd["Temps suppression (ms)"]
temps_suppression_btree = data_btree["Temps Suppression"]
temps_suppression_max_heap = data_max_heap["Temps Suppression"]

# Création des graphiques
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Insertion
axs[0].plot(data_abr["Taille"], data_abr["Temps insertion"], label="ABR", marker='o')
axs[0].plot(data_list_bd["Taille"], data_list_bd["Temps insertion (ms)"], label="List BD", marker='o')
axs[0].plot(data_btree["Degrée"], data_btree["Temps Insertion"], label="Btree (Degrée)", marker='o')
axs[0].plot(data_max_heap["Taille"], data_max_heap["Temps Insertion"], label="Max Heap", marker='o')
axs[0].set_title("Comparaison des temps d'insertion")
axs[0].set_xlabel("Taille / Degrée")
axs[0].set_ylabel("Temps (ms)")
axs[0].set_xscale("log")
axs[0].set_yscale("log")
axs[0].legend()
axs[0].grid(True, which="both", linestyle="--", linewidth=0.5)

# Recherche
axs[1].plot(data_abr["Taille"], temps_recherche_abr, label="ABR", marker='o')
axs[1].plot(data_list_bd["Taille"], temps_recherche_list_bd, label="List BD", marker='o')
axs[1].plot(data_btree["Degrée"], temps_recherche_btree, label="Btree (Degrée)", marker='o')
axs[1].plot(data_max_heap["Taille"], temps_recherche_max_heap, label="Max Heap", marker='o')
axs[1].set_title("Comparaison des temps de recherche")
axs[1].set_xlabel("Taille / Degrée")
axs[1].set_ylabel("Temps (ms)")
axs[1].set_xscale("log")
axs[1].set_yscale("log")
axs[1].legend()
axs[1].grid(True, which="both", linestyle="--", linewidth=0.5)

# Suppression
axs[2].plot(data_abr["Taille"], temps_suppression_abr, label="ABR", marker='o')
axs[2].plot(data_list_bd["Taille"], temps_suppression_list_bd, label="List BD", marker='o')
axs[2].plot(data_btree["Degrée"], temps_suppression_btree, label="Btree (Degrée)", marker='o')
axs[2].plot(data_max_heap["Taille"], temps_suppression_max_heap, label="Max Heap", marker='o')
axs[2].set_title("Comparaison des temps de suppression")
axs[2].set_xlabel("Taille / Degrée")
axs[2].set_ylabel("Temps (ms)")
axs[2].set_xscale("log")
axs[2].set_yscale("log")
axs[2].legend()
axs[2].grid(True, which="both", linestyle="--", linewidth=0.5)

plt.tight_layout()
plt.show()
plt.savefig("complexite_Structures.png")
```
