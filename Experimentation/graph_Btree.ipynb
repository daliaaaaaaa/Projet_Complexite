# Analyse des Temps d'Op√©rations

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

B-tree en C

```python
# Read the CSV file
data = pd.read_csv('complexite_temps_Btree.csv')

plt.figure(figsize=(18, 6))

# Graphique pour la recherche
plt.subplot(1, 3, 1)
plt.plot(data['Degr√©e'], data['Temps Recherche'], marker='o', label="Temps Recherche", color='blue')
plt.xlabel("Degr√©e")
plt.ylabel("Temps (ms)")
plt.title("Temps de Recherche")
plt.xscale("linear")
plt.yscale("linear")
plt.xticks(data['Degr√©e'], rotation=45)
plt.grid()
plt.legend()

# Graphique pour l'insertion
plt.subplot(1, 3, 2)
plt.plot(data['Degr√©e'], data['Temps Insertion'], marker='o', label="Temps Insertion", color='green')
plt.xlabel("Degr√©e")
plt.ylabel("Temps (s)")
plt.title("Temps d'Insertion")
plt.xscale("linear")
plt.yscale("linear")
plt.xticks(data['Degr√©e'], rotation=45)
plt.grid()
plt.legend()

# Graphique pour la suppression
plt.subplot(1, 3, 3)
plt.plot(data['Degr√©e'], data['Temps Suppression'], marker='o', label="Temps Suppression", color='red')
plt.xlabel("Degr√©e")
plt.ylabel("Temps (s)")
plt.title("Temps de Suppression")
plt.xscale("linear")
plt.yscale("linear")
plt.xticks(data['Degr√©e'], rotation=45)
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
