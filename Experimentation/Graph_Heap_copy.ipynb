# Analyse des Temps d'Op√©rations

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Max Heap C

```python
# Lecture du fichier CSV
max_heap = pd.read_csv('./complexite_temps_MAX_HEAP.csv')

plt.figure(figsize=(18, 6))

# Graphique pour la recherche
plt.subplot(1, 3, 1)
plt.plot(max_heap['Taille'], max_heap['Temps Recherche'], marker='o', label="Temps Recherche", color='blue')
plt.plot(max_heap['Taille'], max_heap['Taille'], label="y=x", linestyle='--', color='gray')
plt.xlabel("Taille")
plt.ylabel("Temps (s)")
plt.title("Temps de Recherche")
plt.xscale("log")
plt.yscale("log")
plt.xticks(max_heap['Taille'], rotation=45)
plt.grid()
plt.legend()

# Graphique pour l'insertion
plt.subplot(1, 3, 2)
plt.plot(max_heap['Taille'], max_heap['Temps Insertion'], marker='o', label="Temps Insertion", color='green')
plt.plot(max_heap['Taille'], np.log2(max_heap['Taille']), label="y=log2(n)", linestyle='--', color='orange')
plt.xlabel("Taille")
plt.ylabel("Temps (s)")
plt.title("Temps d'Insertion")
plt.xscale("log")
plt.yscale("log")
plt.xticks(max_heap['Taille'], rotation=45)
plt.grid()
plt.legend()

# Graphique pour la suppression
plt.subplot(1, 3, 3)
plt.plot(max_heap['Taille'], max_heap['Temps Suppression'], marker='o', label="Temps Suppression", color='red')
plt.plot(max_heap['Taille'], np.log2(max_heap['Taille']), label="y=log2(n)", linestyle='--', color='orange')
plt.xlabel("Taille")
plt.ylabel("Temps (s)")
plt.title("Temps de Suppression")
plt.xscale("log")
plt.yscale("log")
plt.xticks(max_heap['Taille'], rotation=45)
plt.grid()
plt.legend()
plt.savefig('Max_Heap.png')
plt.tight_layout()
plt.show()
```

```python
# Cr√©ation du graphique avec 4 courbes pour l'insertion uniquement
plt.figure(figsize=(14, 8))

# Courbe Min-Heap en C++
plt.plot(min_heap_cpp['Taille'], min_heap_cpp['Temps Insertion'], label='Min-Heap C++ Insertion', color='blue', marker='o')

# Courbe Min-Heap en C
plt.plot(min_heap['Taille'], min_heap['Temps Insertion'], label='Min-Heap C Insertion', color='green', marker='o')

# Courbe Max-Heap en C++
plt.plot(max_heap_cpp['Taille'], max_heap_cpp['Temps Insertion'], label='Max-Heap C++ Insertion', color='red', marker='o')

# Courbe Max-Heap en C
plt.plot(max_heap['Taille'], max_heap['Temps Insertion'], label='Max-Heap C Insertion', color='orange', marker='o')

# Ajouter des labels et un titre
plt.xlabel('Taille du Tas')
plt.ylabel('Temps (en secondes)')
plt.title('Comparaison des temps d\'insertion entre Min-Heap et Max-Heap en C et C++')

# Ajouter une grille
plt.grid(True)

# Ajouter une l√©gende
plt.legend()

# Enregistrer l'image
plt.savefig('Comparaison_Insertion_Min_Max_Heap_C_C++.png')

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
# Charger les fichiers CSV dans des DataFrames
resultats_on_c = pd.read_csv('resultats_Tri_ON.csv')
resultats_onlogn_c = pd.read_csv('resultats_Tri_ONLogN.csv')
resultats_on_cpp = pd.read_csv('resultats_ON_C++.csv')
resultats_onlogn_cpp = pd.read_csv('resultats_ONLogN_C++.csv')

# Tracer les r√©sultats pour O(n)
plt.figure(figsize=(10, 6))
plt.plot(resultats_on_c['Taille'], resultats_on_c['Temps de Tri (s)'], label="C O(n)", marker='o')
plt.plot(resultats_on_cpp['Taille'], resultats_on_cpp['Temps de Tri (s)'], label="C++ O(n)", marker='s')
plt.xlabel("Taille du vecteur")
plt.ylabel("Temps de tri (s)")
plt.title("Temps de tri pour O(n)")
plt.legend()
plt.grid(True)
plt.show()

# Tracer les r√©sultats pour O(n log n)
plt.figure(figsize=(10, 6))
plt.plot(resultats_onlogn_c['Taille'], resultats_onlogn_c['Temps de Tri (s)'], label="C O(n log n)", marker='o')
plt.plot(resultats_onlogn_cpp['Taille'], resultats_onlogn_cpp['Temps de Tri (s)'], label="C++ O(n log n)", marker='s')
plt.xlabel("Taille du vecteur")
plt.ylabel("Temps de tri (s)")
plt.title("Temps de tri pour O(n log n)")
plt.legend()
plt.grid(True)
plt.show()

# Tracer tous les r√©sultats sur un seul graphique
plt.figure(figsize=(12, 8))
plt.plot(resultats_on_c['Taille'], resultats_on_c['Temps de Tri (s)'], label="C O(n)", marker='o')
plt.plot(resultats_onlogn_c['Taille'], resultats_onlogn_c['Temps de Tri (s)'], label="C O(n log n)", marker='o')
plt.plot(resultats_on_cpp['Taille'], resultats_on_cpp['Temps de Tri (s)'], label="C++ O(n)", marker='s')
plt.plot(resultats_onlogn_cpp['Taille'], resultats_onlogn_cpp['Temps de Tri (s)'], label="C++ O(n log n)", marker='s')
plt.xlabel("Taille du vecteur")
plt.ylabel("Temps de tri (s)")
plt.title("Comparaison des temps de tri pour O(n) et O(n log n) en C et C++")
plt.legend()
plt.grid(True)
plt.show()
plt.savefig('Comparaison_Tri_TAS_C_C++.png')

```
